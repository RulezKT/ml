```python
def w_sum(a,b):
    assert(len(a) == len(b))
    output = 0
    for i in range(len(a)):
        output += (a[i] * b[i])
    return output

def vect_mat_mul(vect,matrix):
    # print(vect)
    # print(matrix)
    # print(len(vect))
    # print(len(matrix))
    assert(len(vect) == len(matrix))
    output = [0,0,0]
    for i in range(len(vect)):
        output[i] = w_sum(vect,matrix[i])
    return output



def neural_network(input: list, weight: list) -> float:
   return  vect_mat_mul(input, weight)


def outer_prod(vec_a,vec_b):
    out = zeroes_matrix(len(vec_a),len(vec_b))
    for i in range(len(vec_a)):
        for j in range(len(vec_b)):
            out[i][j] = vec_a[i] * vec_b[j]
    return out

def zeroes_matrix(a,b):
    out = []
    for i in range(a):
        out.append([0]*b)
    return out
```
