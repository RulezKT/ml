```python
import random
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

faces = ['H', 'T']
num_sims = 100
num_flips = 20

results = []

for sim in range(num_sims):
    coins = random.choices(faces, k=num_flips)
    results+= [coins.count("H")]

plt.scatter(range(num_sims), results);

plt.xlabel('Experiment number')
plt.ylabel('Number of heads observed')
plt.title('Number of heads observed in 20 coin flips');
```

```python
edges = np.arange(1.5, 16, 1)
#plt.yticks(np.arange(0, 22, 2))
#plt.grid(axis='y')
plt.hist(results, bins = edges, edgecolor='black', density=True)
```

```python
def coinsim_print(num_sims = 100, num_flips = 20,     threshold = 6):

    faces = ['H', 'T'] 

    for sim in range(num_sims):
        coins = random.choices(faces, k=num_flips)
        num_heads = coins.count("H")
        if num_heads <= threshold:
            print(f"Experiment {sim} has {num_heads} heads")
```

```python
coinsim_print()
```

```python
def coinsim(num_sims = 1000, num_flips = 20,     threshold = 6):
    event_count = 0

    for sim in range(num_sims):
        coins = random.choices(faces, k=num_flips)
        num_heads = coins.count("H")
        if num_heads <= threshold:
            event_count += 1
    print(f"relative frequency of {threshold} or fewer is: {event_count/num_sims}")
```

```python
coinsim(1_000_000)
```

```python
coinsim(1_000_000, 20, 4)
```
